```python
# ============================================================================
# SECTION 2: ENCODING CATEGORICAL DATA - COMPREHENSIVE SUMMARY
# ============================================================================
#
# WHY ENCODE?
# Machine learning algorithms work with NUMBERS, not text. We must convert
# categorical data (strings) into numerical representations.
#
# ----------------------------------------------------------------------------
# METHOD 1: MANUAL ENCODING using .map()
# ----------------------------------------------------------------------------
#
# How it works:
# df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
#
# When to use:
#   ✅ Binary categories (2 values)
#   ✅ Small number of categories you want to control
#   ✅ Need to preserve the mapping for interpretation
#   ✅ Ordinal data with inherent order (low/medium/high → 0/1/2)
#
# Pros: Full control, reversible, transparent
# Cons: Tedious for many categories
#
# Example from our practical:
# - Gender: Male=0, Female=1
# - Why 0/1? Mathematical convenience, standardization, no false magnitude
#
# ----------------------------------------------------------------------------
# METHOD 2: CAT.CODES (Automatic Encoding)
# ----------------------------------------------------------------------------
#
# How it works:
# df['State_names'] = df['State_names'].astype('category')
# df['State_names'] = df['State_names'].cat.codes
#
# When to use:
#   ✅ Many categories (10+)
#   ✅ Quick exploratory analysis
#   ✅ Memory efficiency matters
#   ✅ Don't need to reverse the mapping later
#
# Pros: Fast, automatic, memory efficient (uses int8/int16)
# Cons: Alphabetical ordering, mapping lost, implies false ordering
#
# Example from our practical:
# - State_names: 50 states encoded as 0-49 alphabetically
# - Alabama=0, Alaska=1, ..., Wyoming=49
#
# ⚠️ CRITICAL LIMITATION: The mapping is lost! You can't easily reverse it.
# The algorithm also thinks state 49 > state 0, which is meaningless.
#
# ----------------------------------------------------------------------------
# METHOD 3: ONE-HOT ENCODING using pd.get_dummies()
# ----------------------------------------------------------------------------
#
# How it works:
# encoded = pd.get_dummies(df['Segment'], prefix='Segment', dtype=int)
# df = pd.concat([df, encoded], axis=1)
# df = df.drop('Segment', axis=1)
#
# When to use:
#   ✅ NOMINAL DATA (no inherent order): colors, regions, customer segments
#   ✅ Moderate number of categories (2-20)
#   ✅ Need to avoid false ordering between categories
#   ✅ Building ML models (most algorithms prefer this for nominal data)
#
# Pros: No false ordering, interpretable, ML-friendly
# Cons: Creates many columns (curse of dimensionality),
#       not ideal for 100+ categories
#
# What it does:
# Instead of one column with values, creates separate binary columns:
#   Segment          →    Segment_Basic  Segment_Gold  Segment_Silver
#   "Basic"          →           1             0             0
#   "Gold"           →           0             1             0
#   "Silver"         →           0             0             1
#
# Each row has exactly ONE "1" (hence "one-hot") and rest are 0s.
#
# Example from our practical:
# - Segment: 5 categories → 5 binary columns
# - Segment_Basic, Segment_Gold, Segment_Missing, Segment_Platinum, Segment_Silver
#
# ----------------------------------------------------------------------------
# DECISION RULE: WHICH METHOD TO USE?
# ----------------------------------------------------------------------------
#
# | Data Type                              | Method              |
# |----------------------------------------|---------------------|
# | Binary (2 categories)                  | .map() → 0/1        |
# | Ordinal (ordered: low/med/high)        | .map() → 0/1/2      |
# | Nominal (no order: colors, segments)   | One-Hot Encoding    |
# | Many categories (50+ states, IDs)      | cat.codes OR avoid  |
#
# KEY INSIGHT: One-Hot Encoding prevents the algorithm from assuming
# false relationships. When encoding "Red", "Blue", "Green" as 0, 1, 2,
# the algorithm thinks Green(2) > Blue(1) > Red(0), which is meaningless!
#
# ============================================================================
```

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler_standard = StandardScaler()

# Select ONLY the numeric columns we want to scale
# (We don't scale the one-hot encoded columns or ID)
columns_to_scale = ['Age', 'Amount_spent', 'Referal', 'Gender']

# Fit the scaler and transform the data
df_scaled_standard = df.copy()  # Make a copy so we can compare
df_scaled_standard[columns_to_scale] = scaler_standard.fit_transform(df[columns_to_scale])

# Check the results
print("After StandardScaler:")
print(df_scaled_standard[columns_to_scale].describe())
print("\n" + "="*50)
print("Notice: mean ≈ 0, std ≈ 1")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMax scaler
scaler_minmax = MinMaxScaler()

# Scale the same columns
df_scaled_minmax = df.copy()
df_scaled_minmax[columns_to_scale] = scaler_minmax.fit_transform(df[columns_to_scale])

# Check the results
print("After MinMaxScaler:")
print(df_scaled_minmax[columns_to_scale].describe())
print("\n" + "="*50)
print("Notice: min = 0, max = 1")
```

```python
# =============================================================================
# SECTION 3: FEATURE SCALING - NEURAL NETWORKS CLARIFICATION
# =============================================================================

# QUESTION: Should you use StandardScaler or MinMaxScaler for Neural Networks?
# ANSWER: BOTH can work! The choice depends on specific factors.

# -----------------------------------------------------------------------------
# STANDARDSCALER FOR NEURAL NETWORKS - WHEN TO USE:
# -----------------------------------------------------------------------------

# ✓ Hidden layers with activation functions that expect centered data:
#   - Sigmoid (outputs 0 to 1, but works better with centered inputs)
#   - Tanh (outputs -1 to 1, designed for centered data)
#   - Helps with gradient flow during backpropagation

# ✓ You want to preserve the original distribution shape
#   - Maintains relative distances between points
#   - Keeps relationships "true to size"

# Example: StandardScaler with tanh/sigmoid activations
# from sklearn.preprocessing import StandardScaler
# X_scaled = StandardScaler().fit_transform(X)
# model = Sequential([
#     Dense(64, activation='tanh'),    # Works well with centered data
#     Dense(32, activation='sigmoid'),
#     Dense(1, activation='sigmoid')
# ])

# -----------------------------------------------------------------------------
# MINMAXSCALER FOR NEURAL NETWORKS - WHEN TO USE:
# -----------------------------------------------------------------------------

# ✓ Using ReLU activation (most modern networks!):
#   - ReLU (outputs 0 to ∞, naturally expects positive inputs)
#   - MinMaxScaler gives [0, 1] → perfect for ReLU!

# ✓ Input layer expects bounded values:
#   - Image data (pixels 0-255 → 0-1)
#   - Probabilities or percentages

# Example: MinMaxScaler with ReLU activations
# from sklearn.preprocessing import MinMaxScaler
# X_scaled = MinMaxScaler().fit_transform(X)
# model = Sequential([
#     Dense(64, activation='relu'),    # Works great with [0,1] inputs
#     Dense(32, activation='relu'),
#     Dense(1, activation='sigmoid')
# ])

# -----------------------------------------------------------------------------
# THE EXPERT DECISION FRAMEWORK FOR NEURAL NETWORKS:
# -----------------------------------------------------------------------------

# Factor 1: ACTIVATION FUNCTIONS
#   - ReLU/LeakyReLU → MinMaxScaler slightly better
#   - Tanh/Sigmoid → StandardScaler slightly better

# Factor 2: YOUR DATA
#   - Images/bounded data → MinMaxScaler
#   - Features with outliers → StandardScaler

# Factor 3: INDUSTRY PRACTICE (2024+)
#   - Most modern deep learning uses MinMaxScaler or no scaling (for ReLU)
#   - Classic ML/shallow networks often use StandardScaler

# -----------------------------------------------------------------------------
# KEY INSIGHT: "TRUE TO SIZE"
# -----------------------------------------------------------------------------

# Both scalers preserve relationships, just differently:
#   - StandardScaler: preserves standard deviation distances
#   - MinMaxScaler: preserves proportional positions in range

# Neither is "more true" - they're just different transformations!
# What matters is matching the scaler to your algorithm and data.

# =============================================================================
```
